```python
import requests
import json

# ---
# Test 1: Ask a question that should be in the Knowledge Base
# (Run scripts/ingest_math_dataset.py first!)
# ---
kb_question = "Natalia sold 48 cupcakes. If she sold 60% of her cupcakes, how many did she have in total?"

response = requests.post(
    "http://localhost:8000/ask/",
    json={"question": kb_question, "student_id": "notebook_test"}
)

print(f"Status Code: {response.status_code}")
print(json.dumps(response.json(), indent=2))
```

```python
# ---
# Test 2: Ask a question that is NOT in the Knowledge Base
# This will test the Web/MCP fallback (which we simulated to return None)
# ---
web_question = "What is the quadratic formula?"

response = requests.post(
    "http://localhost:8000/ask/",
    json={"question": web_question, "student_id": "notebook_test"}
)

print(f"Status Code: {response.status_code}")
print(json.dumps(response.json(), indent=2))
```
